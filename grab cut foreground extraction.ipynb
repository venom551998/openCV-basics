```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('opencv-python-foreground-extraction-tutorial.jpg')
#cv2.imshow('img',img)
x = img.shape
# print(x)
y = img.shape[:2]
# print(y)
mask = np.zeros(img.shape[:2],np.uint8)
# print(mask.shape)
# print(mask)
bgdModel = np.zeros((1,65),np.float64)
# print(bgdModel.shape)
# print(bgdModel)
fgdModel = np.zeros((1,65),np.float64)
rectangle = (80,50,300,400)
cv2.grabCut(img,mask,rectangle,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==0) | (mask==2) , 0 , 1).astype('uint8')              #0=cv2.GC_BGD    2=cv2.GC_PR_BGD
img = img*mask2[:,:,np.newaxis]
plt.imshow(img)
plt.colorbar()
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
